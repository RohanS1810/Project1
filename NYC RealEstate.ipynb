```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import hvplot.pandas
import geoviews as gv
from geoviews import dim, opts
```

```python
Manhattan_df = pd.read_csv(Path("rollingsales_manhattan.csv"))
Manhattan_df.head()
```

```python
Brooklyndata_df = pd.read_csv(Path("rollingsales_brooklyn.csv"))
Queensdata_df = pd.read_csv(Path("rollingsales_queens.csv"))
statendata_df = pd.read_csv(Path("rollingsales_statenisland.csv"))
```

```python
Bronxdata_df = pd.read_csv(Path("rollingsales_bronx.csv"))
```

```python
#Concat function to add all the NYC neighborhood data
all_neighborhoods_df = pd.concat(
    [Brooklyndata_df, Manhattan_df, Queensdata_df, statendata_df, Bronxdata_df], 
    axis="rows",
    sort=False,
)
all_neighborhoods_df.head()
```

```python
all_neighborhoods_df.columns
#converting SALE PRICE to Numeric 
all_neighborhoods_df[' SALE PRICE '] = all_neighborhoods_df[' SALE PRICE '].str.replace(',','')
all_neighborhoods_df[' SALE PRICE '] = all_neighborhoods_df[' SALE PRICE '].apply(pd.to_numeric)
all_neighborhoods_df
```

```python
#Calculate the highest sales price in each of the 5 city in a year Manhattan Manhattan Bronx Brooklyn Queens Staten island
max_prices = all_neighborhoods_df.groupby('NEIGHBORHOOD')[' SALE PRICE '].max()
max_prices
```

```python
#Calculate the highest sales price in each of the Neighborhood for the 5 Borough's
max_prices.hvplot.bar(
    x='NEIGHBORHOOD',
    y=' SALE PRICE ',
    xlabel='Neighborhood',
    ylabel='Max Sale Price',
    title='Max Sale Prices by Neighborhood',
    height=1000,
    width=1200
).opts(yformatter='%.0f')
```

Max house prices is 931 million in upper east side

```python
#Hvplot of highest sales price in each of the Neighborhood for the 5 Borough's
max_prices.hvplot.bar(
    x="NEIGHBORHOOD",
    xlabel="NEIGHBORHOOD",
    ylabel="SALE PRICE",
    groupby="NEIGHBORHOOD",
    title="Sale Price By Neighborhood"
).opts(yformatter='%.0f')
```

```python
df_filtered = all_neighborhoods_df.loc[all_neighborhoods_df[' SALE PRICE '] > 5000]
```

```python
#2 Calculate the min sales price in each of the 5 city in a year Manhattan Bronx Brooklyn Queens Staten island
min_prices = df_filtered.groupby('NEIGHBORHOOD')[' SALE PRICE '].min()
min_prices
```

```python
#3 Group the data by neighborhood and average sales price for each neighborhood in NYC City
mean_prices = all_neighborhoods_df.groupby('NEIGHBORHOOD')[' SALE PRICE '].mean()
mean_prices
```

```python
# group the data by borough and average sales price for each neighbourhood in NYC City
mean_prices.hvplot.bar(
    x='NEIGHBORHOOD',
    y=' SALE PRICE ',
    xlabel='Neighborhood',
    ylabel='Average Sale Price',
    title='Average Sale Prices by Neighborhood',
    height=1000,
    width=1200
).opts(yformatter='%.0f')
```

```python
#Calculate which Building Class category was sold the most in the 5 nyc boroughs and build a bar plot for  building classes categrory for each borough
```

```python
building_class_counts = all_neighborhoods_df.groupby('BOROUGH')['BUILDING CLASS CATEGORY'].count()
building_class_counts
```

```python
# sort the counts in descending order and select the top 10 categories
top_building_classes = building_class_counts.sort_values(ascending=False).head(10)
top_building_classes
```

```python
# reset the index to turn the Series into a DataFrame
top_building_classes = top_building_classes.reset_index()
top_building_classes
```

```python
listbuildingclass_df = all_neighborhoods_df.groupby(['BOROUGH','BUILDING CLASS CATEGORY']).size()
listbuildingclass_df
buildingclasscategory_df = listbuildingclass_df.to_frame(name = 'size').reset_index()
buildingclasscategory_df
```

```python
#Manhattan = 1, Bronox = 2, Brooklyn = 3, Queens = 4 , Staten island = 5
layout = buildingclasscategory_df.hvplot.bar(
    x='BOROUGH', 
    y='size', 
    by='BUILDING CLASS CATEGORY', 
    #groupby='BUILDING CLASS CATEGORY', 
    stacked=True, 
    cmap='Category20', 
    #legend='top_left',
    width=1500,
    xlabel='',
)
layout
```
